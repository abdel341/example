## Latitue and Longitude of the neighberhoods

```python
lats=pd.read_csv('Geospatial_Coordinates.xls')
lats

DATA_FINAL["Latitude"]=lats["Latitude"]
DATA_FINAL["Longitude"]=lats["Longitude"]
DATA_FINAL=DATA_FINAL.sort_values(by=['Postcode'])

DATA_FINAL = DATA_FINAL.reset_index(drop=True)
DATA_FINAL
```

## clustering the neighbourhoods (I decided to use the whole neighbourhoods instead of removing some)

```python
kclusters = 4


# run k-means clustering
kmeans = KMeans(n_clusters=kclusters, random_state=0).fit(DATA_FINAL[["Latitude", "Longitude"]])
# add clustering labels
#DATA_FINAL.insert(0, 'Cluster Labels', kmeans.labels_)
DATA_FINAL
```

## Creating the map

```python
# create map
map_clusters = folium.Map(location=[latitude, longitude], zoom_start=11)

# set color scheme for the clusters
x = np.arange(kclusters)
ys = [i + x + (i*x)**2 for i in range(kclusters)]
colors_array = cm.rainbow(np.linspace(0, 1, len(ys)))
rainbow = [colors.rgb2hex(i) for i in colors_array]

# add markers to the map
markers_colors = []
for lat, lon, poi, cluster in zip(DATA_FINAL['Latitude'], DATA_FINAL['Longitude'], DATA_FINAL['Neighbourhood'], DATA_FINAL['Cluster Labels']):
    label = folium.Popup(str(poi) + ' Cluster ' + str(cluster), parse_html=True)
    folium.CircleMarker(
        [lat, lon],
        radius=5,
        popup=label,
        color=rainbow[cluster-1],
        fill=True,
        fill_color=rainbow[cluster-1],
        fill_opacity=0.7).add_to(map_clusters)
       
map_clusters
```
